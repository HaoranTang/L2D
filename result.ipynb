```python
import numpy as np
import matplotlib.pyplot as plt
```

SAC MLP

```python
sac_mlp_50 = np.load('logs/sac/Carla-v0_21/sac_50k.npz')
sac_mlp_l_50 = sac_mlp_50['ep_lengths'].mean(axis=1)
sac_mlp_r_50 = sac_mlp_50['results'].mean(axis=1)

sac_mlp_200 = np.load('logs/sac/Carla-v0_24/sac_200k.npz')
sac_mlp_l_200 = sac_mlp_200['ep_lengths'].mean(axis=1)
sac_mlp_r_200 = sac_mlp_200['results'].mean(axis=1)
plt.plot(sac_mlp_r_200[::5], label='200k steps')
plt.plot(sac_mlp_r_50[::5], label='50k steps')
plt.legend()
plt.title('SAC MLP Eval Reward')
plt.show()
```

```python
plt.plot(sac_mlp_l_200[::5], label='200k steps')
plt.plot(sac_mlp_l_50[::5], label='50k steps')
plt.legend()
plt.title('SAC MLP Eval Length')
plt.show()
```

DDPG MLP

```python
ddpg_mlp_50 = np.load('logs/ddpg/Carla-v0_2/50k.npz')
ddpg_mlp_l_50 = ddpg_mlp_50['ep_lengths'].mean(axis=1)
ddpg_mlp_r_50 = ddpg_mlp_50['results'].mean(axis=1)

ddpg_mlp_200 = np.load('logs/ddpg/Carla-v0_5/200k.npz')
ddpg_mlp_l_200 = ddpg_mlp_200['ep_lengths'].mean(axis=1)
ddpg_mlp_r_200 = ddpg_mlp_200['results'].mean(axis=1)

plt.plot(ddpg_mlp_r_200[::5], label='200k steps')
plt.plot(ddpg_mlp_r_50[::5], label='50k steps')
plt.legend()
plt.title('DDPG MLP Eval Reward')
plt.show()
```

```python
plt.plot(ddpg_mlp_l_200[::5], label='200k steps')
plt.plot(ddpg_mlp_l_50[::5], label='50k steps')
plt.legend()
plt.title('DDPG MLP Eval Length')
plt.show()
```
